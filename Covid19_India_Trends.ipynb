```python
#Importing Libraries matplotlib,pandas,numpy and sk learn
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
```

```python
#reading the dataset
df = pd.read_csv(r"C:\Users\Priyanshu Panda\Desktop\Latest Covid-19 India Status.csv")
```

```python
df
```

```python
df.info()
```

```python
#Deleting some collumns
df.drop(['Active Ratio','Death Ratio','Discharge Ratio'],axis = 1,inplace =True)
```

```python
df
```

```python
#finding the number  of unique states
States = df['State/UTs'].unique()
len(States)
```

```python
for idx in range(0,len(States)):
    C = df[df['State/UTs'] == States[idx]].reset_index()
    plt.scatter(np.arange(0,len(C)),C['Total Cases'],color = 'blue',label = 'Confirmed')
    plt.scatter(np.arange(0,len(C)),C['Deaths'],color = 'red',label ='Deaths')
    plt.scatter(np.arange(0,len(C)),C['Discharged'],color='green',label='Recovered')
    plt.title(States[idx])
    plt.xlabel('Current Day')
    plt.ylabel('Number of Cases')
    plt.legend()
    plt.show()
    
```

```python
df2 = df.nlargest(n=10,columns=['Deaths'])
```

```python
df2
```

```python
df3 = pd.DataFrame(df2,columns = ['State/UTs','Deaths'])
```

```python
df3.plot(x = 'State/UTs',y = 'Deaths',kind ='bar',color = 'red')
plt.title("TOP 10 STATES DEATH WISE")
plt.show()
```

```python
df4 = pd.DataFrame(df,columns = ['State/UTs','Active'])
df4.plot(x='State/UTs',y='Active',kind ='bar',color='yellow')
plt.title("Active Cases")
plt.show()
```

```python
df5=pd.DataFrame(df,columns = ['State/UTs','Deaths','Total Cases'])
df5['Case Fatality Rate'] = (df5['Deaths'] / df5['Total Cases'])*100
df6=pd.DataFrame(df5,columns = ['State/UTs','Case Fatality Rate'])
df6.plot(y='Case Fatality Rate',x='State/UTs' ,kind ='bar')
plt.title("Case Fatality Percentage")
```

```python
#Positive Cases Per Capita
df7 = pd.DataFrame(df,columns = ['State/UTs','Population','Total Cases'])
df7['Positive Cases Per Captia'] =(df7['Total Cases']/df7['Population'])
df7.plot(x='State/UTs',y='Positive Cases Per Captia',kind ='bar',color ='orange')
plt.title("Positive Cases Per captia")
plt.show()

```
